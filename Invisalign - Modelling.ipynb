```python
df1=df1[df1['final_text']!=1]
```

```python
### Remove comments less than 3 letter words

def count_words(text):
    x=[i for i in text.split(" ")]
    if len(x) < 5:
        return np.nan
    return text
df1['final_text']=df1['final_text'].apply(count_words)
```

```python
df1.dropna(inplace=True)
```

```python
np.round(100 * (df1['Option 1'].value_counts()/df1['Option 1'].shape[0]),1).astype('str') + '%'
```

```python
## Data Cleaning
import re
def clean_text(text):
    if isinstance(text, str):
    # Remove all non-alphanumeric characters and extra spaces
        cleaned_text = re.sub(r'[^a-zA-Z0-9\s]+', '', text)
        cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
        return cleaned_text.strip()
    else:
        return ''
# Apply the clean_text function to the 'Text' column
df1['Cleaned_Text'] = df1['final_text'].apply(clean_text)
df1
```

```python
# remove stop words

from nltk.corpus import stopwords

# Assuming df1 is your DataFrame with a column 'stemmed_text'
# Create a list of stop words up to three letters
stop_words = stopwords.words('english')
stop_words = [word.lower() for word in stop_words if len(word) > 3]

# Remove stop words from the 'stemmed_text' column
df1['Cleaned_Text']= df1['Cleaned_Text'].str.lower()
df1['Cleaned_Text'] =df1['Cleaned_Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))

# Print the updated DataFrame
df1.head()
```

```python
### Lemmatization
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    words = nltk.word_tokenize(text)
    lemmatized_words = [lemmatizer.lemmatize(word, wordnet.VERB) for word in words]
    return ' '.join(lemmatized_words)

# Apply lemmatization to the text column
df1['lemmatized_text'] = df1['Cleaned_Text'].apply(lemmatize_text)

# Print the DataFrame with lemmatized text
df1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
## TF-IDF Vectorizer

tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the vectorizer on the text column
tfidf_matrix = tfidf_vectorizer.fit_transform(df1['lemmatized_text'])

tfidf_df=pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
df1['id']=[i for i in range(df1.shape[0])]
tfidf_df['id']=[i for i in range(tfidf_df.shape[0])]
tfidf_df=tfidf_df.merge(df1[['id','Option 1']],on=['id'])
```

```python
tfidf_df
```

```python
tfidf_df.isnull().sum()
```

```python
tfidf_df.drop(columns=['id'],inplace=True)
tfidf_df.rename(columns={'Option 1':'labels'},inplace=True)
tfidf_df['labels'].isnull().sum()
```

```python
# Lebel Encoding the labels
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()

# Fit and transform the label encoder on the category column
tfidf_df['labels'] = label_encode.fit_transform(tfidf_df['labels'])

# Print the DataFrame with encoded categories
tfidf_df['labels'].value_counts()
```

```python
### Train test split
from sklearn.model_selection import train_test_split

X = tfidf_df.drop(columns=['labels']) # Features
y = tfidf_df['labels']  # Target

# X_train=X.iloc[:-1000,:]
# y_train=y.iloc[:-1000]
# X_test=X.iloc[-1000:,:]
# y_test=y.iloc[-1000:]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42,shuffle=True)
```

```python
### Modelling Softmax Classification - Baseline

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


softmax_classifier = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000)

# Train the classifier
softmax_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = softmax_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
import warnings
from sklearn.metrics import confusion_matrix, classification_report

warnings.filterwarnings('ignore')
class_names = [i for i in df1['Option 1'].unique().tolist()]
report = classification_report(y_test, y_pred, target_names=class_names)

print("Classification Report:\n", report)
```
